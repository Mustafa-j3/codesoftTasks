```python
import pandas as pd
```

```python
data = pd.read_csv("G:\Analysis_folder\Dataset\\movies.csv",usecols =  ['movieId','title'])
data.head()
```

```python
rating_data = pd.read_csv("G:\Analysis_folder\Dataset\\ratings1.csv", usecols = ['userId','movieId','rating'])
rating_data.head()
```

```python
rating_data.pivot(index='movieId',columns='userId',values='rating')
```

```python
from scipy.sparse import csr_matrix
```

```python
movie_matrix = csr_matrix(rating_data.values)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model = NearestNeighbors(metric = 'cosine',algorithm = 'brute', n_neighbors = 20)
model.fit(movie_matrix)
```

```python
from fuzzywuzzy import process
```

```python
def recommend(movie,x,n):
    ind = process.extractOne(movie,data['title'])[2]
    print('Movie selected', data['title'][ind], 'index: ', ind)
    distance, indices = model.kneighbors(x[ind], n_neighbors=n)
    for i in indices:
        print(data['title'][i].where(i!=ind))
```

```python
recommend('toy story',movie_matrix,10)
```
